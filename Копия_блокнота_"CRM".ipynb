#### **Задача 1**

Создайте класс `Lead`.

**Решение**

Напишите свое решение ниже

```python
from dataclasses import dataclass

@dataclass
class Lead:
    lead_id: int
    datetime: str
    page: str
    form_title: str
    contacts: str
    name: str

    def to_dict(self):
        return self.__dict__
```

✏️ ✏️ ✏️

**Проверка**

Чтобы проверить свое решение, запустите код в следующих ячейках

```python
lead = Lead(1, '2023-07-01 10:00', 'example.com', 'Обратная связь', 'ivan@example.com', 'Иван Иванов')
res = lead.to_dict()

try:
    assert res == {'lead_id': 1,
                   'datetime': '2023-07-01 10:00',
                   'page': 'example.com',
                   'form_title': 'Обратная связь',
                   'contacts': 'ivan@example.com',
                   'name': 'Иван Иванов'}
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```

#### **Задача 2**

Создайте класс `Simulator`.

**Решение**

Напишите свое решение ниже

```python
class Simulator:
    def __init__(self, simulator_type, version):
        self.simulator_type = simulator_type
        self.version = version

    def to_dict(self):
        return {
            'simulator_type': self.simulator_type,
            'version': self.version
        }
```

#### **Задача 4**

К классу `SalesFunnel` добавьте метод `process_lead`.

**Решение**

Напишите свое решение ниже

```python
import requests

class SalesFunnel:
    def __init__(self, stages=None):
        self.stages = stages if stages is not None else []

    def add_stage(self, stage_name):
        self.stages.append(stage_name)

    def remove_stage(self, stage_name):
        self.stages.remove(stage_name)

    def process_lead(self, lead):
        way = []

        current_stage = self.stages[0]  # Начинаем с первого этапа
        way.append(f"Обработка лида {lead.lead_id} - Текущий этап: {current_stage}")

        crm_url = "https://example.com/crm/api/leads"
        crm_params = {
            'lead_id': lead.lead_id,
            'datetime': lead.datetime,
            'page': lead.page,
            'form_title': lead.form_title,
            'contacts': lead.contacts,
            'name': lead.name
        }
        response = requests.post(crm_url, json=crm_params)
        print(response)
        if response.status_code == 200:
            way.append("Данные лида успешно отправлены в CRM")
        else:
            way.append("Не удалось отправить данные лида в CRM")

        # Перемещаем лид по этапам воронки (для примера, просто переходим к следующему этапу)
        next_stage_index = self.stages.index(current_stage) + 1
        if next_stage_index < len(self.stages):
            next_stage = self.stages[next_stage_index]
            way.append(f"Перемещаем лид {lead.lead_id} на следующий этап: {next_stage}")
        else:
            way.append(f"Лид {lead.lead_id} достиг конца воронки")

        return way
```

✏️ ✏️ ✏️

**Проверка**

Чтобы проверить свое решение, запустите код в следующих ячейках

```python
sales_funnel = SalesFunnel(["Генерация лида", "Подтверждение", "Предложение", "Возражение"])
lead = Lead("123", "2023-07-01", "example.com", "Свяжитесь с нами", "ivan@example.com", "Иван Иванов")
res4 = sales_funnel.process_lead(lead)

try:
    assert res4 == ['Обработка лида 123 - Текущий этап: Генерация лида',
                    'Не удалось отправить данные лида в CRM',
                    'Перемещаем лид 123 на следующий этап: Подтверждение']
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```
