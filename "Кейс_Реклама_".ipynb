### **Найдите самую эффективную платформу по средним значениям конверсий**

Cохраните результат в переменную `most_effective_platform`

````python
import csv

with open("campaign_data.csv", mode="r", encoding="utf-8", newline="") as file:
    reader = csv.DictReader(file)

    # Словарь: ключ — название канала, значение — список показателей конверсии
    channel_conversions = {}
    for record in reader:
        channel = record["Платформа"]
        conversion = int(record["Конверсия"])

        if channel not in channel_conversions:
            channel_conversions[channel] = [conversion]
        else:
            channel_conversions[channel].append(conversion)

# Расчёт средних показателей конверсии по каналам продвижения
average_conversions = {
    channel: sum(conversions) / len(conversions)
    for channel, conversions in channel_conversions.items()
}

# Определение наиболее эффективного канала (с максимальной средней конверсией)
most_effective_platform = max(average_conversions, key=average_conversions.get)

print(most_effective_platform)
````

### **Найдите самый успешный тип рекламной кампании по средним значениям конверсий**

Сохраните результат в переменную `most_successful_campaign_type`

````python
import csv

with open("campaign_data.csv", mode="r", encoding="utf-8", newline="") as csv_file:
    reader = csv.DictReader(csv_file)

    # Словарь: ключ — категория кампании, значение — список значений конверсии
    campaign_categories = {}
    for row in reader:
        category = row["Тип кампании"]
        conversion = int(row["Конверсия"])
        if category not in campaign_categories:
            campaign_categories[category] = [conversion]
        else:
            campaign_categories[category].append(conversion)

# Расчёт средних конверсий по каждой категории кампаний
avg_conversions_by_category = {
    category: sum(conversions) / len(conversions)
    for category, conversions in campaign_categories.items()
}

# Определение наиболее эффективной категории кампаний
most_successful_campaign_type = max(
    avg_conversions_by_category.items(),
    key=lambda item: item[1]
)[0]
most_successful_campaign_type
````

### **Найдите самый успешный город проведения кампании по средним значениям конверсий**

Сохраните результат в переменную `most_successful_city`

````python
import csv

with open("campaign_data.csv", mode="r", encoding="utf-8", newline="") as csv_file:
    reader = csv.DictReader(csv_file)

    # Словарь: ключ — город, значение — список значений конверсии
    city_conversions = {}
    for row in reader:
        city = row["Город"]
        conversion = int(row["Конверсия"])
        if city not in city_conversions:
            city_conversions[city] = [conversion]
        else:
            city_conversions[city].append(conversion)

    # Расчёт средней конверсии по каждому городу
    avg_conversions = {
        city: sum(conversions) / len(conversions)
        for city, conversions in city_conversions.items()
    }

    # Определение города с максимальной средней конверсией
    most_successful_city = max(avg_conversions.items(), key=lambda x: x[1])[0]
    print (most_successful_city)
````

### **Посчитайте долю затрат на рекламные кампании каждой платформы и каждого города**

Сохраните результат в файл `platform_city_results.txt`

Результат должен выглядеть примерно так:

```
Для группы Telegram:
- Город: Москва, доля затрат на рекламу: 16.46%
- Город: Калуга, доля затрат на рекламу: 16.2%
- Город: Казань, доля затрат на рекламу: 14.57%
- ...
Для группы Вконтакте:
- Город: Екатеринбург, доля затрат на рекламу: 17.43%
- Город: Новосибирск, доля затрат на рекламу: 14.86%
- Город: Санкт-Петербург, доля затрат на рекламу: 14.13%
...
```

Города внутри групп платформ должны быть отсотированы по убыванию доли затрат, а доля затрат должны быть округлена до 2 знаков после запятой.

````python
import csv

# Загружаем данные из CSV‑файла в список словарей
with open("campaign_data.csv", mode="r", encoding="utf-8", newline="") as file:
    reader = csv.DictReader(file)
    records = [row for row in reader]


analysis_results = {}

for row in records:
    # Извлекаем значения из записи
    campaign_id = row["ID Кампании"]
    start_date = row["Начальная дата"]
    end_date = row["Конечная дата"]
    campaign_type = row["Тип кампании"]
    platform = row["Платформа"]
    budget = row["Бюджет"]
    city = row["Город"]
    shows = row["Показы"]
    clicks = row["Клики"]
    conversion = row["Конверсия"]
    revenue = row["Доход"]

    # Используем платформу как ключ для группировки данных
    key = platform

    # Инициализируем структуру для новой платформы
    if key not in analysis_results:
        analysis_results[key] = {"cities": {}, "total_budget": 0}

    # Добавляем город в структуру, если его ещё нет
    if city not in analysis_results[key]["cities"]:
        analysis_results[key]["cities"][city] = 0

    # Преобразуем бюджет в число и добавляем к общей сумме
    budget_value = round(float(budget))
    analysis_results[key]["total_budget"] += budget_value

    # Увеличиваем сумму затрат для конкретного города
    analysis_results[key]["cities"][city] += budget_value

# Сохраняем результаты в текстовый файл
with open("platform_city_results.txt", mode="w", encoding="utf-8") as output_file:
    # Обрабатываем каждую платформу
    for platform_key in analysis_results:
        city_data = analysis_results[platform_key]["cities"]
        total_spent = analysis_results[platform_key]["total_budget"]

        # Сортируем города по убыванию затрат
        sorted_cities = sorted(city_data.items(), key=lambda item: item[1], reverse=True)

        # Формируем вывод для файла
        output_file.write(f"Для группы {platform_key}:\n")
        for city_name, city_budget in sorted_cities:
            budget_share = round(city_budget / total_spent * 100, 2)
            output_file.write(f"- Город: {city_name}, доля затрат на рекламу: {budget_share}%\n")
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python
from functools import reduce

def calculate_average_budget(path, platform):
    with open(path, mode="r", encoding="utf-8", newline='') as f:
      reader = csv.DictReader(f)
      data = [row for row in reader]

    filtered_data = filter(lambda campaign: campaign['Платформа'] == platform, data)
    budget_list = list(map(lambda campaign: float(campaign['Бюджет']), filtered_data))
    return round(reduce(lambda a, b: a + b, budget_list) / len(budget_list), 2)

calculate_average_budget('campaign_data.csv', 'Google')
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python
from datetime import datetime, timedelta, date

def get_missing_campaign_dates(path):
    with open(path, mode="r", encoding="utf-8", newline='') as f:
        reader = csv.DictReader(f)
        data = [row for row in reader]

    date_list = [campaign['Начальная дата'] for campaign in data]
    unique_dates = sorted(set(date_list), key=lambda x: datetime.strptime(x, '%Y-%m-%d').date())
    start_date = datetime.strptime(unique_dates[0], '%Y-%m-%d').date()
    end_date = date(2022, 12, 30)
    date_range = (start_date + timedelta(days=x) for x in range((end_date - start_date).days))
    campaign_dates = set(datetime.strptime(d, '%Y-%m-%d').date() for d in date_list)

    for d in date_range:
        if d not in campaign_dates:
            yield d
````

### **Напишите функцию, которая будет принимать на вход путь к файлу и возвращать сгруппированные данные по городам, содержащие суммарный бюджет и количество кликов**

Функция `group_campaign_data` должна возвращать список словарей, отсортированный по городам, в формате:

```python
[{'Город': 'Нижний Новгород',
  'Количество кликов': 616.0,
  'Суммарный бюджет': 27585},
 {'Город': 'Новосибирск',
  'Количество кликов': 733.0,
  'Суммарный бюджет': 33001},
  ...]
```

````python
from operator import itemgetter
from itertools import groupby

def group_campaign_data(path):
    with open(path, mode="r", encoding="utf-8", newline='') as f:
      reader = csv.DictReader(f)
      data = [row for row in reader]

    sorted_data = sorted(data, key=itemgetter('Город'))

    grouped_data = []

    for city, campaigns in groupby(sorted_data, key=itemgetter('Город')):
        clicks_total = 0
        budget_total = 0
        for campaign in campaigns:
            clicks_total += float(campaign.get('Клики', 0))
            budget_total += int(campaign.get('Бюджет', 0))
        grouped_data.append({'Город': city, 'Количество кликов': clicks_total, 'Суммарный бюджет': budget_total})

    return grouped_data

group_campaign_data('campaign_data.csv')
````

### **Напишите генератор, который будет генерировать данные о кампаниях, запущенных в указанном городе и имеющих бюджет выше заданного значения**

Функция `campaign_generator` должна принимать путь к файлу, город и бюджет и генерировать словари, отсортированные по возрастанию 'ID Кампании' в формате:

```
{'ID Кампании': '657', 'Тип': 'Контекстная', 'Платформа': 'Вконтакте', 'Доход': '10692'}
{'ID Кампании': '674', 'Тип': 'Нативная интеграция', 'Платформа': 'Яндекс', 'Доход': '20766'}
{'ID Кампании': '761', 'Тип': 'Соцсети', 'Платформа': 'Вконтакте', 'Доход': '15680'}
```

````python
def campaign_generator(path, city, budget):
    with open(path, mode="r", encoding="utf-8", newline='') as f:
      reader = csv.DictReader(f)
      data = [row for row in reader]

    filtered_data = filter(lambda campaign: campaign['Город'] == city and int(campaign['Бюджет']) > budget, data)

    campaigns = ({'ID Кампании': campaign['ID Кампании'],
                  'Тип': campaign['Тип кампании'],
                  'Платформа': campaign['Платформа'],
                  'Доход': campaign['Доход']} for campaign in filtered_data)

    return campaigns

campaigns = campaign_generator('campaign_data.csv', 'Москва', 25000)

for campaign in campaigns:
    print(campaign)
````
